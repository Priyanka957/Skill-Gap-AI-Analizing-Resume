Question 13:Print how many days are left for your next birthday.

```python
import datetime 
def calNextBday(month, day) :
    today = datetime.date.today()
    myBday = datetime.date(today.year, month, day)
    daysLeft = 0
    if today > myBday :
        daysLeft = (datetime.date(myBday.year + 1, myBday.month, myBday.day) - today).days
    elif today < myBday :
        daysLeft = (myBday - today).days
    return daysLeft
year = int(input("Enter your birthday year : "))
month = int(input("Enter your Birthday month : "))
day = int(input("Enter your Birthday date : "))
daysLeft = calNextBday(month, day)
print("Your Birthday date :",datetime.date(year, month, day) )
print("Days remaining for your next Birthday :",daysLeft)
```

Question 14:Convert the date '2022-05-15' to a datetime object and add 30 days.

```python
import datetime as dt
text = '2022-012-29'
lst = list(map(int, text.split('-')))
extraDays = dt.timedelta(days=30)
givenDate = dt.date(lst[0], lst[1], lst[2])
newDate = givenDate + extraDays
print(givenDate)
print(newDate)
```
