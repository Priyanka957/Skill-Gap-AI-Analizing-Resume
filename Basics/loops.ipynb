Question 1:Print numbers from 1 to 10 using a for loop

```python
for i in range(1, 11) :
    print(i, end = ' ')
```

Question 2:Print all even numbers from 1 to 20

```python
for i in range(1,21):
    if i % 2 == 0 :
        print(i, end = ' ')
```

Question 4:Using a while loop, print numbers from 5 down to 1.

```python
n = 5
while n > 0 :
    print(n, end = ' ')
    n -= 1
```

Question 5:Write a loop to find the sum of numbers from 1 to 50

Question 7:Find how many vowels are present in the string 'Programming'.

```python
text = 'Programming'
vowelCount = 0
for i in text.lower() :
    if i in 'aeiou' :
        vowelCount += 1
print('Number of vowels in Programming : ',vowelCount)
```

Question 8:Use a loop to reverse the string 'PythonLoops'

```python
print("Reverse of 'PythonLoops':")
s = "PythonLoops"
rev = ""
for ch in s:
    rev = ch + rev
print(rev)
print()
```

Question 9:Print numbers from 1–10, but skip 5 using continue.

```python
for i in range(1, 11) :
    if i == 5 :
        continue
    print(i, end = ' ')
```

Question 10:Print numbers from 1–20, but stop when number reaches 13 using break.

```python
for i in range(1, 21) :
    if i == 13 :
        break
    print(i, end = ' ')
```

Question 11:Write a loop to check if a number is prime

```python
import math
n = int(input("Enter a number : "))
prime = True
for i in range(2, int(math.sqrt(n))+1) :
    if n % i == 0 :
        prime = False
print(n, "is prime" if prime else "is not prime")
```

Question 12:Count how many times each character occurs in 'mississippi'.

```python
print("Character count in 'mississippi':")
word = "mississippi"
count_dict = {}

for ch in word:
    if ch in count_dict:
        count_dict[ch] += 1
    else:
        count_dict[ch] = 1

print(count_dict)
print()
```
