Question 3:Write a function square(n) that returns the square of n.

```python
def square(n):
    return n * n

print("3. Square of 5 =", square(5))
print()
```

Question 4:Write a function calculator(a, b) that returns the sum, difference, and product of a & b

```python
def calculator(a, b):
    return a + b, a - b, a * b

s, d, p = calculator(10, 4)
print("4. Sum:", s)
print("Difference:", d)
print("Product:", p)
print()
```

Question 6:Write a function total(*nums) that returns the sum of all numbers passed.

```python
def total(*nums):
    return sum(nums)

print("6. Total =", total(5, 10, 15))
print()
```

Question 7:Write a function student_info(**data) that prints key : value for all items.

```python
def student_info(**data):
    for key, value in data.items():
        print("7.",key, ":", value)

student_info(name="Priya", age=20, course="Python")
print()
```

Question 8:Write a function count_vowels(text) that returns the number of vowels in the string

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for ch in text:
        if ch in vowels:
            count += 1
    return count

print("8. Vowels in 'Programming' =", count_vowels("Programming"))
print()
```

Question 9: Write a lambda function to return the cube of a number

```python
cube = lambda x: x * x * x
print("9. Cube of 3 =", cube(3))
print()
```

Question 10:Write a function unique_letters(text) that returns unique letters in the order they appear.

```python
def unique_letters(text):
    result = ""
    for ch in text:
        if ch not in result:
            result += ch
    return result

print("10. Unique letters:", unique_letters("mississippi"))
```
