Question 1: Create a CSV file named students.csv with columns name, age, grade and write 3 student records.

```python
import csv
with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["name", "age", "grade"])
    writer.writerow(["Priyanka", 22, "A"])
    writer.writerow(["Rahul", 23, "B"])
    writer.writerow(["Anita", 21, "A"])
print("students.csv created")
```

Question 2:Write a Python program to read and print each row from employees.csv.

Question 3:From products.csv, print only the first column (product names).

Question 7:Read sales.csv and calculate total sales (item, price, quantity).

```python
import csv

total = 0
with open("D:/Skill-Gap-AI-Analizing-Resume/csv/q7sales.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        total += int(row["price"]) * int(row["quantity"])
print("Total Sales:", total)
```

Question 8:Convert a list of lists into a CSV file (id, name).

```python
import csv

data = [
    [1, "Priyanka"],
    [2, "Rahul"],
    [3, "Anita"]
]

with open("D:/Skill-Gap-AI-Analizing-Resume/csv/q8list_data.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["id", "name"])
    writer.writerows(data)
print("CSV created from list")
```

Question 9:Filter and print rows from people.csv where age > 30.

```python
import csv

with open("D:/Skill-Gap-AI-Analizing-Resume/csv/q9people.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        if int(row["age"]) > 30:
            print(row)
```

Question 10:Copy all contents from source.csv to copy.csv.

```python
import csv

with open("D:/Skill-Gap-AI-Analizing-Resume/csv/q10source.csv", "r") as src, open("copy.csv", "w", newline="") as dest:
    reader = csv.reader(src)
    writer = csv.writer(dest)
    for row in reader:
        writer.writerow(row)
print("File copied successfully")
```
