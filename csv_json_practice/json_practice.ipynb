Question 1: Convert the JSON string {'name': 'Alex', 'age': 22, 'city': 'Chennai'} into a Python dictionary.

Question 2: Convert the dictionary {'id': 101, 'product': 'Laptop', 'price': 55000} into a JSON string.

```python
import json
data = {'id': 101, 'product': 'Laptop', 'price': 55000}
json_str = json.dumps(data)
print(json_str)
```

Question 5: From JSON {'employee': {'name': 'John', 'age': 30, 'department': 'HR'}}, print only the department.

```python
import json
employee = {'employee': {'name': 'John', 'age': 30, 'department': 'HR'}}
print(employee['employee']['department'])
```

Question 8: From JSON {'users': [{'id':1,'name':'Alex'}, {'id':2,'name':'Mia'}]}, print all user names.

Question 9: Add a new key 'country': 'India' to {'name': 'Sam', 'age': 28} and convert back to JSON.

```python
import json
content = {"name":"Sam","age":28}
print("Before adding :",content)
content["country"] = "India"
content = json.dumps(content, indent=4)
print("After adding :",content)
```

Question 10: Create items.json with [{'item':'Pen','price':10}, {'item':'Book','price':50}, {'item':'Bag','price':700}] and print total cost.

```python
import json
items = [
    {'item': 'Pen', 'price': 10},
    {'item': 'Book', 'price': 50},
    {'item': 'Bag', 'price': 700}
]
with open("items.json", "w") as file:
    json.dump(items, file)
total = sum(item['price'] for item in items)
print("Total cost:", total)
```
